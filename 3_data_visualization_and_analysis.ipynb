# üß† Notebook 3 ‚Äî SpaceX Data Visualization & Insights

```
# Cell 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the cleaned dataset
df = pd.read_csv("cleaned_spacex_api.csv")
print("‚úÖ Data loaded successfully.")
print("Shape:", df.shape)
df.head()
```

```
# Cell 2
df['launch_year'] = pd.to_datetime(df['date_utc']).dt.year

yearly_success = df.groupby('launch_year')['success'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=yearly_success, x='launch_year', y='success', marker='o', color='tab:blue')
plt.title("üöÄ SpaceX Launch Success Rate Over Time")
plt.ylabel("Success Rate")
plt.xlabel("Launch Year")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```
# Cell 3
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='rocket', y='success', palette='coolwarm', ci=None)
plt.title("üõ∞Ô∏è Launch Success Rate by Rocket Type")
plt.ylabel("Average Success Rate")
plt.xlabel("Rocket Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# Cell 4
fig = px.bar(
    df, 
    x='launch_year', 
    color='success',
    title="üìÖ SpaceX Launches Over the Years",
    labels={'launch_year':'Year', 'success':'Success Status'},
    barmode='group'
)
fig.show()
```

```
# Cell 5
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='viridis', fmt='.2f')
plt.title("üìà Correlation Between Numeric Features")
plt.show()
```

```
# Cell 6
df.to_csv("spacex_visualized.csv", index=False)
print("‚úÖ Visualization data saved successfully as 'spacex_visualized.csv'")
```
